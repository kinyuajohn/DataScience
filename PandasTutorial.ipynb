# Pandas Intro

```python
import pandas as pd

# Read csv using pandas
# df -> DataFrame

# Read excel
# df_xlsx = pd.read_excel('modified_pokemon_data.xlsx')
# df_xlsx

# Read text file
# df_txt = pd.read_csv('modified_pokemon_data.txt', delimiter='\t')
# df_txt

# Read csv
df = pd.read_csv('pokemon_data.csv')
df.head()
```

# Reading Data in Pandas

```python
# Read Headers
print(df.columns)
```

```python
# Read each columns
df[['Name', 'Type 1', 'HP']][0:5]
```

```python
df.describe()
```

```python
# SOrting
df.sort_values('Name', ascending=False)
```

```python
# Sorting 2
df.sort_values(['Type 1', 'HP'], ascending=[1, 0])
```

# Making Changes to the Data

```python
df.head(5)
```

```python
# Add a column
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
# Drop a column(s)
df = df.drop(columns = ['Total'])
df.head(5)
```

```python
# Add a column (different method)
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.head(5)
```

# Saving our Data

```python
# Save as CSV
df.to_csv('modified_pokemon_data.csv', index=False)
```
